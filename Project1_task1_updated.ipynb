
Cluster for data1-8(kmeans)

In this problem, we don't need to plot an elbow plot to find the best n_cluster parameter. This is because the original dataset already contains a true class column. I just find the number of true classes to represent the n_cluster

```python
##dataset with 3 features
##set random_states = 0 since kmeans is non deterministic
##hierarchy clustering is deterministic 
from sklearn.cluster import KMeans
def kmean_3d(dataset):
    n_clusters = len(dataset.Class.unique())
    mdl = KMeans(n_clusters = n_clusters, random_state = 0)
    mdl.fit(dataset.iloc[:,0:3])
    return mdl.inertia_, mdl.cluster_centers_, mdl.labels_+1
```

```python
from sklearn.cluster import KMeans
def kmean_2d(dataset):
    n_clusters = len(dataset.Class.unique())
    mdl = KMeans(n_clusters = n_clusters, random_state = 0)
    mdl.fit(dataset.iloc[:,0:2])
    return mdl.inertia_, mdl.cluster_centers_, mdl.labels_+1
```

```python
##cluster for data1--kmeans
d1_inertia_, d1_cluster_centers_, d1_labels_ = kmean_3d(data1)
d1_labels_ 
data1['predictedclass_kmeans'] = d1_labels_
data1
```

```python
##define function to implement hierarchy clustering algo(with 3 features)
from sklearn.cluster import AgglomerativeClustering
def hierarchy_3d(dataset):
    n_clusters = len(dataset.Class.unique())
    mdl = AgglomerativeClustering(n_clusters = n_clusters, linkage = 'ward', 
                                  ).fit(dataset.iloc[:,0:3])
    return mdl.labels_+1
```

```python
from sklearn.cluster import AgglomerativeClustering
def hierarchy_2d(dataset):
    n_clusters = len(dataset.Class.unique())
    mdl = AgglomerativeClustering(n_clusters = n_clusters, linkage = 'ward',
                                  ).fit(dataset.iloc[:,0:2])
    return mdl.labels_+1
```

2. Evaluate the performance of clustering using external validation.

From the table obtained above, we noticed that labels assigned to predicteclass_kmeans and predictedclass_hierarchy are not semantic and do not matched with original label. Therefore we would have to map original label with these two predicted classes and obtain the correct jaccard_score. 

```python
# data1
# Class 1, predictedclass_kmeans 4, predictedclass_hierarchy 7
# Class 2, predictedclass_kmeans 6, predictedclass_hierarchy 5
# Class 3, predictedclass_kmeans 7, predictedclass_hierarchy 3
# Class 4, predictedclass_kmeans 5, predictedclass_hierarchy 6
# Class 5, predictedclass_kmeans 3, predictedclass_hierarchy 2
# Class 6, predictedclass_kmeans 1, predictedclass_hierarchy 1
# Class 7, predictedclass_kmeans 2, predictedclass_hierarchy 4
data1[data1['Class']==5].head() #need to examine this for each label in data1
```

```python
# Map Data 1 Cluster numbers with true class labels
# Map from A to B
data1['predictedclass_kmeans'] = data1['predictedclass_kmeans'].map({4:1,6:2,7:3,5:4,3:5,1:6,2:7}) 
data1['predictedclass_hierarchy'] = data1['predictedclass_hierarchy'].map({7:1,5:2,3:3,6:4,2:5,1:6,4:7})
```

```python
##view result of mapping
data1.head()
```

```python
##use jaccard_score for external validation to examine performance of kmeans
import sklearn.metrics as metrics
js_kmeans_1 = metrics.jaccard_score(data1['predictedclass_kmeans'], data1['Class'], average = 'macro')
js_kmeans_1
```

```python
##examine performance of hierachy
js_hierarchy_1 = metrics.jaccard_score(data1['predictedclass_hierarchy'], data1['Class'], 
                                       average = 'macro')
js_hierarchy_1
```

Jaccard score measures the similarity of two sets to see which are shared and which are distinct, with a range of 0 to 1. Jaccard score of 1 indicates that clusters obtained with kmeans and hierarchy are the same as original label.
