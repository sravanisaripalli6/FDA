Question 3 : Conduct a facility and county level analysis on shortage of aides, clinical staff, nursing staff and other staff. Define the following metric for this analysis: number of weeks the shortage was experienced. The metric should be categorized into 3 categories, 0-10, 11-20 and >20 weeks. Sample output tables are show below (total number of tables = 4*2 = 8) (20 points) **bold text**

```python
df_n=df_covid.loc[((df_covid['Shortage of Nursing Staff'] == "Y"))] 
n=df_n.groupby(["Provider Name"])["Week Ending"].count().reset_index(name="Week count")
n.loc[n['Week count'] <= 10, 'Number of Shortage weeks'] = '0-10'
n.loc[(n['Week count'] > 10) & (n['Week count'] <=20), 'Number of Shortage weeks'] = '11-20'
n.loc[n['Week count'] > 20, 'Number of Shortage weeks'] = '>20'
n[['Provider Name','Number of Shortage weeks']]
```

**Question 5 : For Middlesex county in Massachusetts compute the hospital bed occupancy in percentage for each facility. How many facilities had 100 % of hospital bed occupancy? (10 points)**

```python
df_covid['Percentage'] = df_covid['Total Number of Occupied Beds']/df_covid['Number of All Beds']*100
beds=df_covid.loc[((df_covid.County == "Middlesex") & (df_covid.Percentage == 100) & (df_covid['Provider State']=="MA"))]
beds['Provider Name'].nunique()
```

**Question 6: From question 5 list all the facilities that had filled to 100% capacity for three weeks or more (10 points)**

```python
c1=beds.groupby(['Provider Name','County'],as_index=False)['Week Ending'].count()
c1.loc[c1['Week Ending']>=3,['Provider Name','Week Ending']]
```
